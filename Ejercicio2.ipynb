```python
# Cargar el dataset proporcionado por el usuario
file_path = 'netflix_titles.csv'
netflix_data = pd.read_csv(file_path)

dataset_info = netflix_data.info()
dataset_description = netflix_data.describe(include='all')
missing_values = netflix_data.isnull().sum()

content_type_distribution = netflix_data['type'].value_counts()

movie_durations = netflix_data[netflix_data['type'] == 'Movie']['duration'].str.replace(' min', '').astype(float)

country_distribution = netflix_data['country'].value_counts().head(10)

release_year_distribution = netflix_data['release_year'].value_counts().sort_index()

print("Dataset Information:")
print(dataset_info)
print("\nDataset Description:")
print(dataset_description)
print("\nMissing Values:")
print(missing_values)
print("\nContent Type Distribution:")
print(content_type_distribution)
print("\nCountry Distribution (Top 10):")
print(country_distribution)
print("\nRelease Year Distribution:")
print(release_year_distribution)

# Generar gráficos
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

# Gráfico de distribución de tipos de contenido
axs[0].bar(content_type_distribution.index, content_type_distribution.values, color='lightblue')
axs[0].set_title('Distribution of Content Types')
axs[0].set_xlabel('Type')
axs[0].set_ylabel('Count')

# Gráfico de distribución de duración de películas
axs[1].hist(movie_durations, bins=30, color='lightgreen')
axs[1].set_title('Distribution of Movie Durations')
axs[1].set_xlabel('Duration (minutes)')
axs[1].set_ylabel('Count')

# Gráfico de distribución de contenido por país (Top 10)
axs[2].bar(country_distribution.index, country_distribution.values, color='lightcoral')
axs[2].set_title('Distribution of Content by Country (Top 10)')
axs[2].set_xlabel('Country')
axs[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
